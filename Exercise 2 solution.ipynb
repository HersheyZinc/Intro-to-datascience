```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df = df.select_dtypes(include='int64')
df = df.drop(['Id','YearBuilt','YrSold','MSSubClass','OverallQual','OverallCond'], axis=1)
df
```

```python
df.describe()
```

```python
sb.set(style="darkgrid", color_codes=True)
```

```python
sales = df['SalePrice']
f, axes = plt.subplots(1, 1, figsize=(24,4))
sb.boxplot(data=sales, orient='h')


sb.displot(sales, aspect=5)


f, axes = plt.subplots(1, 1, figsize=(24,4))
sb.kdeplot(sales)

```

```python
lotarea = df['LotArea']
f, axes = plt.subplots(1, 1, figsize=(24,4))
sb.boxplot(data=lotarea, orient='h')


sb.displot(lotarea, aspect=4)


f, axes = plt.subplots(1, 1, figsize=(24,4))
sb.kdeplot(lotarea)
```

```python

sb.jointplot(x=lotarea, y=sales, color='green')
```
